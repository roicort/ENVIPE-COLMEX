```python
import pandas as pd
import plotly.express as px
from tqdm import tqdm

encoding = 'latin1'

# DataFrame para acumular todos los años
df_todos = pd.DataFrame()

for year in tqdm(range(2018, 2025), desc='Procesando años'):
    TPer_Vic1 = pd.read_csv(f'./data/{year}/TPer_Vic1.csv', index_col=0, encoding=encoding)
    Frontera_Tijuana = TPer_Vic1[TPer_Vic1['AREAM'] == 21].copy()
    PercepcionXSexo = Frontera_Tijuana[['AP4_3_2', 'SEXO']].copy()
    PercepcionXSexo['AP4_3_2'] = PercepcionXSexo['AP4_3_2'].replace({1: 'Seguro', 2: 'Inseguro', 9: 'No sabe/No contestó'})
    PercepcionXSexo['SEXO'] = PercepcionXSexo['SEXO'].replace({1: 'Hombre', 2: 'Mujer'})
    PercepcionXSexo['AÑO'] = year
    df_todos = pd.concat([df_todos, PercepcionXSexo], ignore_index=True)

# Graficar todos los años juntos
fig = px.histogram(
    df_todos,
    x='AP4_3_2',
    color='SEXO',
    barmode='group',
    facet_col='AÑO',
    category_orders={"AÑO": sorted(df_todos['AÑO'].unique())},
    labels={'AP4_3_2': 'Percepción de seguridad', 'SEXO': 'Sexo', 'AÑO': 'Año'},
    title='Percepción de seguridad pública en Tijuana por sexo y año'
    )
fig.update_layout(bargap=0.2)
fig.show()
```
